```
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```
base_url = "https://loginmegastore.com/search?query=kulkas&p=true&page="

data_list = []

for page in range(1, 6):
    url = base_url + str(page)
    print(f"Scraping halaman {page}...")

    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Ambil semua produk
    titles = soup.find_all("p", class_="mantine-focus-auto !leading-[120%] m_b6d8b162 mantine-Text-root")
    prices = soup.find_all("p", class_="mantine-focus-auto !whitespace-nowrap m_b6d8b162 mantine-Text-root")
    images = soup.find_all("img", class_="m_9e117634 mantine-Image-root")

    for i in range(len(titles)):
        try:
            title = titles[i].text.strip()
            price = prices[i].text.strip()
            image = images[i]['src']

            data_list.append({
                "title": title,
                "price": price,
                "image": image
            })
        except:
            continue
```
